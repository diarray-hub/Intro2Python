L'encapsulation est un principe qui consiste √† cacher ou prot√©ger certaines donn√©es de notre objet. Dans la plupart des langages orient√©s objet, tels que le C++, le Java ou le PHP, on va consid√©rer que nos attributs d'objets ne doivent pas √™tre accessibles depuis l'ext√©rieur de la classe.

On va d√©finir des m√©thodes un peu particuli√®res, appel√©es des accesseurs et mutateurs.

L'inconv√©nient de devoir √©crire des accesseurs et mutateurs, comme vous l'aurez sans doute compris, c'est qu'il faut cr√©er deux m√©thodes pour chaque attribut de notre classe. Pour faire respecter l'encapsulation propre au langage, on la fonde sur des conventions que nous allons d√©couvrir un peu plus bas mais surtout sur le bon sens de l'utilisateur de notre classe (√† savoir, si j'ai √©crit que cet attribut est inaccessible depuis l'ext√©rieur de la classe, je ne vais pas chercher √† y acc√©der depuis l'ext√©rieur de la classe).
