## Pseudo formattage en utilisant les propri√©t√©s des strs et de la fonction print!
Avant de parler des m√©thodes de formattage en elles m√™me, parlons de quelques syntaxs qu'on a d√©j√† vu ici mais qu'on ne peut pas r√©√©llement consid√©rer comme du formattage! Par exemple sachant qu'on peut utiliser l'operateur "+" pour concatener deux strs on peut utiliser cela pour int√©grer des strings entre eux au moment de l'affichage, on peut aussi modifier les propri√©t√©s d'un afficage en modifiant la valeur des arguments de la fonctions print!

```python
num = 666
name = "Abdou"
solde = 12345789.0
# Par d√©faut l'argument sep de la fonction print va permettre d'afficher ces arguments s√©par√©s par des espaces
# Il est important de noter que pour la fonction print, chaque valeur affich√©e ici est un argument positionnel √† part
print("Compte num√©ro:", num, "; propri√©taire:", name, "; solde:", solde)
# On peut √©galement utiliser '+' pour concatener les strings comme dit pr√©c√©demment
# Cette voici du point de vue de la fonction il n'y a plus qu'un seul argument
print("Compte num√©ro: " + str(num) + "; propri√©taire: " + name + "; solde: " + str(solde))
```

## Affichage format√©e avec l'op√©rateur "%"
Une m√©thode r√©lativement peu utilis√© depuis Python3 mais plut√¥t populaire avant et consid√©r√© comme la seule m√©thode d'affichage format√©e dans d'autres langages comme le C par exemple, on peut utiliser l'op√©rateur "%" suivi d'un suffix variables en fonctions du type de l'objet qu'on souhaite afficher pour int√©grer des ints (%d), floats (%f) et strs (%s)

```python
print("Compte num√©ro: %d"%num)
formatting = (num, name, solde)
print("Compte num√©ro: %d; propri√©taire: %s; solde: %.1f" %formatting)
# C'est le r√¥le de l'op√©rateur % losrqu'il est de cette mani√®re avec les strings
# Notez qu'il est important d'avoir les abstractions "%x" dans le string √† formater
# On peut m√™me les utiliser comme op√©rateur de convertions
convert = "Compte num√©ro: %f" %num
print(convert)
# Disons que je veux limiter cet affichage deux chiffres apr√®s la virgule
print("Compte num√©ro %.1f" %num)
```

## Strings methods!
Les strs eux m√™me en python fournissent des methodes pour faire de l'affichage format√©e, jettons un petit coup d'oeil aux methodes **format, center, ljust and rjust**!
La m√©thode **format** fournit un autre moyen d'int√©grer des objets de diff√©rents type dans un string, les autres permettent d'ajuster la position d'un message par rapport √† un prompt!

```python
print("Compte num√©ro: {}; propri√©taire: {}; solde: {}".format(num, name, solde))
# On peut configurer un systems de cl√©s pour cette m√©thode
print("Compte num√©ro: {key1}; propri√©taire: {key}; solde: {key0}".format(key0=solde, key1=num, key=name))
# Positionner un message proportionnellement √† un prompt
message = "Abdou"
print(message.center(11, "$"))
print(message.ljust(10, "$"))
print(message.rjust(10, "$"))
```

## Formatted strings!
Les **F-strings ou String format√©es** sont un type de str permattant de rendre la syntax des m√©thode comme format, center et etc plus concise et compr√©hensible, disponible depuis la version 3.6 de python, cette m√©thode est largement devenu la plus populaire d√®s lors gr√¢ce √† sa syntax plus naturelle pour les d√©butants!<br/>
Pour cr√©er un formatted string (string formatt√©es) il suffit de le prefixer du carat√®re "f". Notez que les strings formatt√©es ne sont pas un nouveau type de donn√©es, ce sont aussi des str!

```python
fstr = f"Compte num√©ro: {num + 12}; propri√©taire: {name}; solde: {solde}"
print(fstr)
#ljust()
print(f"{'$'*10}{message}")
#rjust
print(f"{message}{'$'*10}")
# center
print(f"{'$'*10}{message}{'$'*10}")
```
