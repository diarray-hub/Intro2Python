## Comment √©crire sa propre fonction?
Les notion pr√©c√©dente peuvent √™tre g√©n√©railiser. **C'est √† dire qu'une fonction peut ne pas avoir d'arguments, tout comme elle peut en avoir un infinit√©. Si elle a des arguments alors ceux ci ont des types, si elle retourne quelque chose alors cette valeur a un type sinon c'est None et les arguments peuvent √™tre utilis√© comme entr√©es dans une fonction, tout comme ils peuvent servir pour changer le comportement de la fonction**<br/>
Maintenant, on arrive √† la plus grosse question donc, sachant tout √ßa maintenant comment on cr√©e des fonctions? Et bien en python il existe deux mots cl√©s permettant de cr√©er des fonctions et leur syntax comme tout en python est simple et direct. Nous avons d'abord la m√©thode la plus classique avec le **keyword def** qui est la mani√®re conventionnelle de cr√©er des fonctions. Pour cela il faut juste pr√©ciser le nom de la fonction apr√®s le mot cl√© et mettre des paranth√®ses dans lesquelles on peut mettre les arguments de la fonctions si elle en prend, puis on mets deux point, on indente et commence √† √©crire le block de code de la fonction. Par exemple voil√† une fonction qui fera le traditionnel Hello world √† chaque fois qu'on appelera:
```
def func():
  print("Hello world")
```
Vous remarquez que avec cette m√©thode on peut √©crire des long block de code complexe pour d√©finir notre fonction et m√™me inclure des bloucles et des structures conditionnelles.<br/>
La deuxi√®me m√©thode beaucoup, moins utilis√©e, est cela dit parfaite pour d√©finir des fonction pouvant t√©nir en une ligne comme celle au dessus, en restant concis et direct. Elle est avec le mot-cl√© **lambda**, pour d√©finir une fonction avec lambda il suffit de mettre les arguments de la fonction (si elle en a) s√©par√©s par des virgules, puis on mets deux points ":" et nous d√©finissons ce que la fonction doit retourner. On peut mettre ce genre de fonction dans une variable qui leur servira de nom:
```

```



````python
# Definis une fonction avec lambda qui retourne la racine carr√© d'un nombre
racine = lambda x: x**0.5

# Definir une fonction qui accueil un utilisateur en fonction de sont nom et de son sexe
def welcome(name, sexe):
  match sexe:
    case "male":
      print("Bienvenu Mr", name)
    case "female":
      print("Bienvenue Mrs", name)
    case _:
      print("Bienvenu(e) cher", name)
````

## Return!
Tr√®s g√©n√©ralement les fonctions retourne des valeurs, **des OBJETS**. D'ailleurs c'est cela qui est abr√©g√© **dans la syntax de lambda, c'est comme si apr√®s les deux points on a un retour implicite**. On peut permettre √† une fonction de faire un retour en utilisant le mot cl√© return √† la fin de la fonction et je pr√©cise bien √† la fin car **le return est la close de fin d'une fonction** Aucun code apr√®s un return ne sera execut√©, c'est la finalit√© des fonctions, si une fonction poss√®de un return, elle s'arr√™tera juste apr√®s avoir ex√©cut√© cette ligne

````python
# Definissons une fonction qui peut nous dire si oui ou non un nombre donn√© est paire
def is_pair(nombre):
  if nombre % 2:
    # si nombre % 2 est diff√©rent de z√©ro cette condition sera vrai et donc on pourra en conclure que le nombre est impaire
    return False
  return True # Si le block pr√©c√©dent est execut√© celle ci ne le sera pas, donc pas vraiment besoin de faire un "else"
````

````python
print(is_pair(100), type(is_pair(100)))
````

## Signature d'une fonction!
De mani√®re simple la signature d'une est l'ensemble des d√©tails qui lui son propre et qui permettent √† l'interpreteur ainsi qu'aux programmer de la reconna√Ætre, **son nom, le nombre d'arguments qu'elle poss√®de, le type de ses arguments, son type de retour** constitue ce qu'on appelle la signature compl√®te d'une fonction. Bien que python √©tant un langage √† faible typage nous autorise √† cr√©er des fonctions avec des signature incompl√®te, dont les types des arguments et de retour notament seront consid√©r√© comme **"Any"** par l'interpreteur, **une signature compl√®te apporte plus de clart√© sur le r√¥le d'une fonction, permet de la distinguer plus facilement d'autre fonction qui pourrait avoir le m√™me nom et permettrait aux utilisateur de savoir quel type de donn√©es doivent ils fournir pour tel et tel argument et de savoir lesquelles sont obligatoires ou optionnels**<br/>
On va redefinir les fonctions pr√©cendente en completant leur signature

````python
# Definir une fonction qui accueil un utilisateur en fonction de sont nom et de son sexe
def welcome(name: str, sexe: str ="male") -> None: # Avec cette nouvelle syntax on pr√©cise le type des arguments et le type de retour
  match sexe:                                      # De plus l'argument "sexe" a maintenant une valeur par d√©faut et devient optionnel
    case "male":
      print("Bienvenu Mr", name)
    case "female":
      print("Bienvenue Mrs", name)
    case _:
      print("Bienvenu(e) cher", name)

# Definissons une fonction qui peut nous dire si oui ou non un nombre donn√© est paire
def is_pair(nombre: int | float) -> bool: # Depuis la version 3.10 de python cette syntax peut √™tre utilis√© pour pr√©ciser le type d'une variable quand cette derni√®re peut en avoir plusieurs
  if nombre % 2:
    # si nombre % 2 est diff√©rent de z√©ro cette condition sera vrai et donc on pourra en conclure que le nombre est impaire
    return False
  return True # Si le block pr√©c√©dent est execut√© celle ci ne le sera pas, donc pas vraiment besoin de faire un "else"
````

````python
welcome("Abdou Sanogo")
print(is_pair(34.4))
````

## Mapping!
Il arrive quelques fois que nous voulions appliquer la m√™me fonction sur une collection de donn√©es sans que cette **fonction n'ait √©t√© design√© pour travailler avec des collections et on ne veut naturellement pas avoir √† r√©ecrire la fonction**. On pourrait faire √ßa avec des boucles mais il s'av√®re que le code deviendrait de plus en plus complexe algorithmiquement avec √ßa. **C'est la qu'intervient le concept de mapping**, pour ce genre de √ßa python a impl√©ment√© une classe "**map**" pour ce genre de cas sp√©cifiquement. Cette classe cr√©e des objets maps contenant l'application de la fonction √† un ensemble de donn√©es. **Let's check this out!**

````python
carte = map(is_pair, (10, -47, -48, 50, 77, 66, 101))
print(carte) # Va afficher l'adrresse de l'objet map cr√©er
# les maps sont des iterables pour avoir acc√®s √† leur contenu individuellement nous devons les parcourir
for resultat in carte: print(resultat, end=" ")
print() # Va √† la ligne

# On peut appliquer des maps √† des fonctions qui prennent plusieurs entr√©es, il suffit de faire autant iterables en entr√©es que d'arguments
# Par d√©faut map s'arr√™tera lorsque le plus petit it√©rable sera satisfait
carte = map(welcome, ("Abdou", "Yacouba", "A√Øcha", "Moussa"), ("male", "male", "female"))
# Pour autant la map contient bien tous les None que la fonction a retourn√©
for resultat in carte: print(resultat)
````

## Generateur!
La grosse question: **c'est quoi un g√©n√©rateur?**<br/>
Vous vous rappelez, tout √† l'heure on disait que la close de fin d'une fonction c'√©tait le **return** et donc qu'une fonction meurt apr√®s avoir fait une retour donc qu'elle ne pouvait toujours retourn√© qu'une seule valeur. Mais et si j'ai besoin de retourner plusieurs valeur, je fais comment?<br/>
Et bien une premi√®re approche serait de retourner une collection dans ce cas, si vous voulez retourner plusieurs valeurs vous pouvez attendre de rassembler toutes les valeurs dans un iterable et retourner cette collection √† la fin de la fonction.<br/>
**Hourra alors, probl√®me r√©gl√©!**<br/>
**Enfa√Æte, non probl√®me pas vraiment r√©solu**, maintenant supposons que pour le bon fonctionement de votre code, votre fameuse fonction doit retourner **un objet par s√©conde** au minimun, si vous devez fournir 10 objets et qu'il faut 50 centi√®mes de s√©conde pour preparer un seul objet, **√ßa fera 5 s√©condes pour que votre fonction puisse retourner la collection!**.<br/>
Donc **comment faire?** Et bien c'est √† √ßa que servent **les G√©n√©rateurs**, m√™me si effectivement le mot cl√© return signifie explicitement fin de la fonction, il existe un autre mot cl√© permettant √† une fonction de retourner un valeur et de continuer son √©x√©cution, **le keyword yield**! C'est donc lui notre sauveur, utiliser **yield** dans une fonction transforme implicitement le type de retour de la fonction en "**Generator**" qui est un autre type d'objet python. A un niveau avanc√© de Python on peut utiliser plein de syntax qui cr√©e implicitement des g√©n√©rateurs. Voyons voir √† quoi √ßa ressemble tout √ßa!

````python
def generate_int(range: range):
  for i in range:
    yield i
    print(i, "a √©t√© retourn√©!")
  print("Tous les √©l√©ments ont √©t√© retourn√©")
````

````python
my_range = range(50, 110, 10)
generated = generate_int(my_range)
print(generated) # Affichera l'adrresse de l'objet "generator cr√©e"
print(*generated) # Va d√©baller l'objet "generator" et l'afficher
````

## Python packing and unpacking method!!
Mais quelle est donc cette **syntax bizarre** qu'on vient de faire:
```
print(*generated)
```
**Qu'est ce que √ßa peut bien signifier?**<br/>
Ne vous inquitez pas j'ai la r√©ponse, figurez vous que l'operateur "*" en python peut aussi √™tre utilis√© pour **emballer et d√©baller** des objets! Je vous explique, on avait dit tout √† l'heure que **la fonction print accepte des arguments positionnel** ce qui veut dire qu'on peut lui donner autant de valeur qu'on le veut si on ne pr√©cise pas √† quel arguments appartient tel valeur, elle cherchera √† afficher la valeur donn√©e. Mais attendez **autant de valeurs qu'on veut?** Il y'a quelque chose de bizarre non? Maintenant qu'on sait d√©finir des fonctions une des premi√®re chose qu'on remarque c'est que **une fonction √† un nombre finie d'arguments**. Comment ont ils reussi un tel exploit lorsqu'ils ont cr√©e la fonction print? Comment peut elle accepter une infinit√© d'arguments positionnels?<br/>
**Well, rien de magique!** Effectivement on ne peut pas d√©clarer une infinit√© d'arguments dans la syntax d'une fonction et c'est l√† qu'il devient utiles **d'emballer les arguments**, c'est l√† que l'operateur "*" fait dans la syntax d'une fonction, lorsqu'il est utilis√© dans la d√©claration de la fonction il dit √† l'interpr√™teur de rassembler toutes les valeurs (qui ne sont pas explicitement assigner √† un argument particulier) donn√©es en arguments lors de l'appel de la fonction dans un iterable qu'on pourra utiliser dans la fonction **sous le nom de l'argument qui est pr√©c√©d√© par l'asterisk**. Et lorsqu'il est utilis√© dans l'appel de la fonction, il dit √† l'interpr√™teur **de d√©baller un objet d√©ballable comme un iterable ou un g√©n√©rateur** et d'utiliser chaque √©l√©ment comme un argument positionnel! Pour mieux visualiser cet affaire bizarre, faisons un exemple avec la fonction is_pair de tout √† l'heure, on va la modifier pour qu'elle puisse accepter une infinit√© d'arguments aussi!


````python
def are_pair(*nombres) -> list[bool]:
    # Example de syntax qui cr√©e un g√©n√©rateur implicitement
    # Cette syntax aussi app√©l√©e "list compression" va cr√©e un g√©n√©rateur qui va retourner si le nombre est pair pour chaque nombre dans l'iterable "nombres"
    # Jusqu'au dernier nombre et il en fera une liste
    response = [is_pair(num) for num in nombres]
    return response
````

````python
results = are_pair(1, 14, 74, 15, 81, 99, 100, 7524)
print(results)
# Ou encore
nombres = (1, 14, 74, 15, 81, 99, 100, 7524)
print(are_pair(*nombres)) # Va d√©baller le tuple "nombres" et utilis√© chaque √©l√©ment comme un argument positionnel comme en haut

# D√©ballage d'une liste
premier, deuxieme, troisieme, *reste = results
print(premier, deuxieme, troisieme)
print(reste)
````

## Variable locale et variable globale!
Maintenant qu'on si conna√Æt un peu en fonction, on va pouvoir expliquer le concept de variable locale et globale. Ce qu'il faut comprendre c'est que de la m√™me mani√®re que les variables n'ont pas de r√©elle identit√© en python, **leur existance est r√©latifs au fichier ou √† la fonction ou ils sont d√©finis**.<br/>
Je m'explique, contraitrement aux fonctions, pseudo constantes et aux classes, les variables ne peuvent pas √™tre import√© d'un module √† un autre en python (On parlera de module juste apr√®s dans la prochaine session), ni d'une fonction √† une autre! **Ce qui veut dire que m√™me dans le m√™me module, vous ne pouvez pas utiliser une variable d√©clar√©e dans une fonction si vous √™tes √† l'exterieur de cette derni√®re**. Les variables n'existent que dans la fonction ou elles sont cr√©es.<br/>
Dans certains cas on peut avoir besoin d'une variable pour la lire et la modifier dans une autre fonction ou dans un autre module, **c'est l√† que les variables globales entre en sc√®ne**. D√©clarer une variable comme √©tant globale vous permet de la modifier dans plusieurs fonctions et modules tant que c'est derniers sont dans le m√™me namespace. Let's check this out!

````python
# Cette cellule va donner une erreur comme quoi: local variable 'var' referenced before assignment
var = "hello"

def func():
  print("La variable globale est: ", var)
  var = "modifi√© par func"
  print("Contenu de la variable globale chang√©!")

func()
````

````python
# Cette cellule va afficher le contenu de la variable globale "var" avant de la changer
global var_globale
var_globale = "hello"
def func():
  global var_globale
  print("La variable globale est: ", var_globale)
  var_globale = "modifi√© par func"
  print("Contenu de la variable globale chang√©!")

func()
print("Nouvelle valeur de la variable globale est: ", var_globale)
````
