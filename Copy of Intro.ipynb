## Switch case in python! (optionnelle)
Donc il y'a un autre concept pas si important dont je tenais √† parler. C'est les statements **switch/case** qui existent dans √©normement de langages de programmation mais qui n'avait pas √©quivalent python jusqu'√† la version 3.10 de python justement. Mais pour √™tre complet quand on parle de structures conditionnelle dans la programmation en g√©n√©rale, il faut en parler!<br/>
Pour faire simple les switch/case sont juste une autre mani√®re de faire des stuctures conditionnelle optimis√© pour les cas la condition est bas√©e sur une variable √† valeur discr√®te et finie. Par exemple, supposons qu'on a une variable qui peut prendre que trois valeurs: "jour", "soir" et "nuit". basiquement un switch/case statement sert √† ce genre de cas, on peut √©videmment faire √ßa avec des if et else if aussi mais le dans la plupart des langages le switch/case est cod√© et compil√© pour contenir moins d'instruction que les structures classiques avec ce genre de cas, donc il sera plus efficace algorithmiquement parlant mais est-ce aussi cas en python? Question! üòÇ M√™me si certains affirment qu'il serait environ **86**% plus efficace √ßa reste relatif. On parlera encore de √ßa surement si on le temps de parl√© de complexit√© algorithmique. Mais en bref voil√† comment √ßa pr√©senterait en C:

```
switch (expression)
‚Äã{
    case constant1:
      // statements
      break;

    case constant2:
      // statements
      break;
    .
    .
    .
    default:
      // default statements
}
```
En python depuis la version 3.10, le statement √† un √©quivalent le **match/case**, m√™me si beaucoup se demandent les d√©v√©loppeurs n'ont pas juste fait pour r√©pondre au d√©mandes des pythonistas car contrairement √† tous les autres mots cl√©s python **match** et **case** sont toujous utilisable comme nom de variable, donc ils sont en quelque sorte des **pseudo keywords** mais **never mind, √ßa fait notre affaire**

````python
time = "jou4r"
match time:
  case "jour":
    print("Bonjour")
  case "soir":
    print("Bonsoir")
  case "nuit":
    print("Bonne nuit")
  case _:   # En python le default case correspont √† underscore
    print("???")
````
